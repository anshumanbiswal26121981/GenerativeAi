### 2.3 What is an LLM?



> "LLM stands for Large Language Model. It's the engine behind ChatGPT, Claude, Gemini, and most of the tools we'll use in this course. Here's what you need to know — and nothing more."

**Large Language Model (LLM):** An AI system trained on massive amounts of text to predict the next word (or "token") in a sequence. You give it text, it continues the text.

*Example:* You type "The capital of France is..." and the LLM predicts "Paris" because it has seen this pattern millions of times in its training data.

That's the entire core idea. You give it text, it continues the text.

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/194/281/original/2.png?1777875226" width="500">

**Why "Large"?**
- Billions of internal parameters (think of them as adjustable dials)
- Trained on internet-scale text (books, websites, code, articles)
- "Large" is what makes them capable of handling such a wide range of tasks

**Parameters** are the internal numbers that the model adjusts during training to get better at predicting text. Think of them like the billions of tiny knobs on a mixing board — each one tuned just right to produce the best output.

**The major LLMs you'll see in this course:**

**One thing to know for later:** Modern LLMs aren't text-only anymore. They can also process images, audio, and video (this is called "multimodal"). But the core mechanism — predict the next token — is still text prediction.

A **token** is a small chunk of text that a model reads and predicts step by step.
Tokens can be whole words, parts of words, punctuation, or even spaces depending on how the text is split. We will study about them in depth ahead.

**Multimodal** means the model can work with more than one type of input or output. Instead of just text-in, text-out, a multimodal model can look at an image and describe it, or listen to audio and transcribe it.

### 3.3 JSON Basics

**JSON (JavaScript Object Notation):** A lightweight format for storing and exchanging data using key-value pairs inside curly braces. It's how computers share structured information with each other.

*Example:* Think of JSON like a labeled filing cabinet. Each drawer has a label (the "key") and contents inside (the "value"). `{"name": "Priya", "age": 24}` — "name" is the label, "Priya" is what's inside.

**JSON = key-value pairs inside curly braces.**

#### Code:

```python
# JSON looks like this:
# (This is a Python dictionary — same syntax as JSON)

# Simple JSON:
user = {
    "name": "Priya",
    "age": 24,
    "is_student": True
}

print("Name:", user["name"])
print("Age:", user["age"])
```

#### Code:

```python
# JSON can be NESTED (objects inside objects):

api_response = {
    "status": "success",
    "data": {
        "message": "Hello! How can I help you today?",
        "model": "gemini-2.5-flash",
        "tokens_used": 42
    },
    "choices": [
        {"text": "Option A"},
        {"text": "Option B"}
    ]
}

# Reading nested values:
print("Message:", api_response["data"]["message"])
print("Model:",   api_response["data"]["model"])
print("First choice:", api_response["choices"][0]["text"])
```

### Part A: Tokens — The Atoms of AI Language


> "LLMs don't read words. They read **tokens**. A token is a chunk of text — sometimes a word, sometimes part of a word, sometimes just a character.
>
> **The analogy: LEGO bricks.**
>
> When you write 'I love programming,' you see 3 words. The LLM sees something like: `[I]` `[ love]` `[ program]` `[ming]` — 4 tokens. It breaks text into LEGO-sized pieces it can work with.
>
> Why should you care? Because **tokens equal money and limits.** Every API call costs per token. Every model has a maximum number of tokens it can handle at once. If you're wasteful with tokens, you're burning cash and hitting walls.
>
> Think of it like data packs from Jio or Airtel — every prompt you send consumes a portion of your monthly allocation, and once you hit the limit, the service slows or stops. Tokens are your AI data pack.

### Part B: Context Window — The Model's Working Memory

> "Okay, so the model reads tokens. But how many tokens can it handle at once? That's the **context window**.
>
> **The analogy: Think of a desk.**
>
> Your context window is the desk size. Everything the model can 'see' — your prompt, the conversation history, AND its response — has to fit on this desk. Small desk? You can only work with a few pages. Huge desk? You can spread out an entire codebase.
>
> When you 'run out of context,' it's like your papers falling off the edge of the desk. The model literally forgets what was at the beginning of your conversation.
>
> Real-world example: ever had a long ChatGPT conversation where it starts contradicting what it said 30 messages ago? That's the desk getting full. Early messages fell off."

**Context Window:** The maximum number of tokens (input + output combined) that a model can process in a single interaction. Everything — your prompt, the conversation history, any documents you paste in, AND the model's response — must fit inside this window.

*Example:* If a model has a 128K context window, think of it as a desk that can hold 128,000 LEGO bricks. If your conversation uses 127,000 bricks, the model only has 1,000 left for its response — and the earliest messages may start "falling off" the desk.

**Context windows in 2026:**

| OpenAI GPT-5.4                 | 272K standard (≈1M opt-in)      | ~700 pages standard / ~2,500 pages at 1M |
| Anthropic Claude Opus 4.6      | 200K standard (1M beta)         | ~500 pages / ~2,500 pages at 1M          |
| Anthropic Claude Sonnet 4.6    | 200K standard (1M beta)         | ~500 pages / ~2,500 pages at 1M          |

**Real-world impact:** Some engineering teams use Gemini's 1M-token context to feed entire microservice codebases into a single prompt and ask architectural questions. With GPT-4o's 128K, the same team has to be selective about which files they include. Neither is wrong — it's about knowing your desk size and working within it.

### Part C: Temperature — The Creativity Dial

> "Last concept — and this one's fun.
>
> **Temperature** controls how creative or predictable the model is. It's a number, usually between 0 and 1.
>
> **The analogy: A chef making a dish.**
>
> Temperature 0 = A chef following a recipe EXACTLY. Measure every gram. Same dish every time. Great for code generation — you want THE correct answer, not a creative interpretation of correctness.
>
> Temperature 1 = A jazz chef. Improvising. 'Let's throw in some sriracha and see what happens.' More creative, more varied, but occasionally... inedible.
>
> For software engineering, you almost always want low temperature. You want deterministic, reliable code — not a model getting 'creative' with your database queries.
>
> Real-world example: GitHub Copilot runs at very low temperature by default. When autocompleting code, you want predictability. But tools like Jasper.ai for marketing copy run at higher temperatures because variety is the point."

**Temperature:** A parameter (usually 0 to 1) that controls the randomness of the model's output. Low temperature means the model picks the most likely next word almost every time (predictable). High temperature means it's willing to pick less likely words (creative, but riskier).

*Example:* Ask "What's 2+2?" at temperature 0 and you always get "4." Ask "Write me a poem about rain" at temperature 0.9 and you'll get a different poem each time — some beautiful, some bizarre.

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/189/426/original/8.png?1776078178" width="500">

**When to use which temperature:**

| Temperature | Use Case | Why |
|-------------|----------|-----|
| **0.0** | Code generation, bug fixes, SQL queries | You want THE correct answer, deterministically |
| **0.1 - 0.3** | Technical docs, test cases, refactoring | Mostly precise, tiny variation for natural language |
| **0.5 - 0.7** | Brainstorming, design discussions, naming | Balance of focus and creativity |
| **0.8 - 1.0** | Creative writing, exploring ideas, generating alternatives | Maximum variety |

**Beginner Note:** Most chat interfaces (ChatGPT, Claude.ai) don't let you set temperature directly — they pick a sensible default. You'll control temperature when calling APIs, which we'll do in later classes.

### See it in Action
Play with the <span style="background-color: blue;">[Temperature & Top-K Visualizer](https://andreban.github.io/temperature-topk-visualizer/)</span> to see how turning the dial changes the mathematical probabilities of the next word.

**Top-K** limits the model to choosing from only the K most likely next tokens instead of all possible tokens.
For example, with Top-K = 5, the model can only pick from the 5 highest-probability next words.

**Top-p** (also called nucleus sampling) limits the model to choosing from the smallest set of tokens whose cumulative probability adds up to **p**.

>Instead of fixing the number of candidate tokens (like Top-K), Top-p fixes the total probability mass.

> For code, you want that consistency of temperature 0. For brainstorming, you want the variety of 0.7+. Knowing which dial to turn is a skill you'll develop throughout this course."

## Choosing the Right Model for the Job

> **Not all models cost the same.** Behind every AI platform, there are multiple model tiers — frontier models that are powerful but expensive, and smaller models that are faster and cheaper but less capable.
>

### The Model Tiers You Need to Know

**Model Tier** refers to the capability level of an LLM. Just like cloud servers come in small, medium, and large, LLMs come in lightweight, mid-range, and frontier variants — each with different capability, speed, and pricing.

| Tier            | Examples                                      | Quality                                   | Speed    | Cost           |
| --------------- | --------------------------------------------- | ----------------------------------------- | -------- | -------------- |
| **Frontier**    | GPT-5.4, Claude Sonnet 4.6, Gemini 3.1 Pro    | Highest — deep reasoning, subtle analysis | Moderate | Highest        |
| **Mid-range**   | GPT-5 mini, Claude Haiku 4, Gemini Flash 3    | Good — handles most everyday tasks well   | Fast     | 10–20x cheaper |
| **Lightweight** | GPT-4o mini, Gemini Nano, smaller open models | Basic — simple tasks, formatting, lookups | Fastest  | 5–10x cheaper  |


When you use ChatGPT's free tier, the platform decides which model to route your request to. When you use the API, **you choose the model explicitly** — and you pay per token based on that choice.

**Real-world parallel:** This is the same tiering you see everywhere in infrastructure — S3 Standard vs S3 Glacier for storage, dedicated vs shared CPU for compute. The capability is different, the price is different, and the right choice depends on the job.

### Why the Same Tokens Cost Different Amounts


> "Remember tokens from earlier? Every API call costs money per token — both input and output. But here's the key insight most people miss: **different models charge wildly different rates for processing the same tokens.**
>
> A 500-token prompt sent to GPT-4o costs roughly 15 times more than the same 500 tokens sent to GPT-4o mini. The tokens are identical. The processing power behind them is not.
>

**Input Tokens vs Output Tokens:**

1. **Input Tokens (Prompt Tokens):** Everything you send *to* the model — your prompt, system instructions, any context or files you attach.
2. **Output Tokens (Completion Tokens):** Everything the model generates and sends *back* to you.

**Why does this matter?** Output tokens require significantly more computational power to generate than input tokens do to read. Because of this, **output tokens almost always cost 2x to 4x more than input tokens.** This is why reigning in a verbose model saves you serious money in production.

### Choosing the Right Model for the Task

> "So how do you decide which tier to use? The answer is the same as choosing a cloud instance: match the capability to the workload."

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/190/803/original/img_%281%29.png?1776327267" width="400">

**The 80/20 rule in practice:** In most developer workflows, roughly 80% of AI interactions are medium-complexity tasks — "explain this error," "write a test for this function," "summarize this PR." A mid-range model handles these well. Save frontier models for the 20% that actually needs deep reasoning.

**When you'll use this:** On the free chat tiers you're using today, the platform handles model routing for you. But starting later when we begin calling LLM APIs directly, you'll choose the model on every request — and this framework will directly determine your costs.

### Token Optimization — Getting Better Output for Fewer Tokens

Beyond choosing the right model, **how you write the prompt** directly affects how many tokens you use — and therefore your cost, speed, and rate limits.

| Habit | Why It Helps | Example |
|-------|-------------|----------|
| **Be specific in your prompt** | Vague prompts produce long, rambling outputs | "List 3 risks" vs "Tell me about risks" (the second generates 3x more output tokens) |
| **Set a length constraint** | Prevents the model from over-explaining | "Answer in under 100 words" or "Maximum 5 bullet points" |
| **Ask for structured output** | Tables and lists use fewer tokens than paragraphs | A 5-row table uses ~60% of the tokens of the same info written as prose |
| **Don't repeat context unnecessarily** | Every token in the input costs money | In long conversations, summarize previous context instead of re-pasting everything |
| **Request only what you need** | Output tokens cost 2-4x more than input tokens | "Return only the function" vs "Explain your approach and then write the function" |

**Advanced Note:** In production systems, teams often use a "model router" — a lightweight classifier that reads the incoming request and routes simple tasks to cheap models and complex tasks to expensive ones. Same pattern as a CDN routing static assets to edge servers and dynamic requests to origin.

> First, **not all models are priced equally.** Frontier models like GPT-5 cost 15-20x more per token than mid-range models like GPT-4o. Same tokens in, very different bill.
>
> Second, **the right model depends on the task complexity.** Simple tasks don't need a frontier model. Complex reasoning tasks justify the cost. This is cloud right-sizing applied to AI.
>
> Third, **how you write the prompt directly affects token usage.** A specific, constrained prompt produces shorter, more focused output — which means fewer tokens, lower cost, and faster responses.
>
> On the free tiers you're using today, this is about speed and rate limits. When you start building with APIs, it's about money. Either way, these habits are worth building now."
